```python
import pandas as pd
```

Opening file from the path in read mode and loading words of file and storing as set because set contains unique elements so duplicate lines will be removed

```python
def get_words(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return {line.strip() for line in file}  #  we are are going to read the file and store it as a set after removing whitespaces
```

Loading tweets by specifying the file path

```python
def load_tweets(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        tweets = file.readlines()  # now we are going to make sure that it read all lines 
    return [tweet.strip() for tweet in tweets] # using list comprehension to iterate over every tweets and removing any whitespaces
```

Storing the provided files 

```python
main_file = 'trump.txt'
pos_file = 'positive.txt'  
neg_file = 'negative.txt'
stopwords_file = 'stopwords.txt'
```

Storing the content of files

```python
twitter = load_tweets(main_file)
positive_words = get_words(pos_file)
negative_words = get_words(neg_file) 
stop_words = get_words(stopwords_file)
```

Removing Special character from files

```python
def clean_tweet(tweet):
    # Splitting the tweet in words
    words = tweet.split()
    
    # Creating a list to store cleaned words
    cleaned_words = []
    
    for word in words:
        # Removing URL
        if word.startswith(('http://', 'https://', 'www.')):
            continue
        
        # Removing special characters and converting it to lower case
        cleaned_word = ''.join(char for char in word if char.isalpha() or char.isspace()).lower()
        
        if cleaned_word:  # Add only non-empty cleaned words
            cleaned_words.append(cleaned_word)
    
    # Join the cleaned words back into a single string
    return ' '.join(cleaned_words)
```

```python
cleaned_tweets = []
for tweet in twitter:
    cleaned_tweets.append(clean_tweet(tweet))
```

Creating varaibles to store the count of pos, neg, stop and other words

```python
overall_positivewords = 0
overall_negativewords = 0
overall_stop_count = 0  
overall_other_count = 0
```

Now getting the words from Cleaned tweets and increasing the count of pos, neg and stop by 1 accordingly

```python
for cleaned_tweet in cleaned_tweets:
    words = cleaned_tweet.split()
    positive_count = 0
    negative_count = 0
    stop_count = 0
    
    for word in words: #iterating through words and updating the count of pos,neg and stop by 1 accordingly
        if word in positive_words:
            positive_count += 1
        elif word in negative_words:
            negative_count += 1
        elif word in stop_words:
            stop_count += 1
    
    other_count = len(words) - (positive_count + negative_count + stop_count) #subtracting the tolat words from the rest of the words to get Other words
```

Calculating the final pos, neg, stop and other count values

```python
total_positive_count=overall_positivewords+positive_count
total_negative_count=overall_negativewords+negative_count
total_stop_count=overall_stop_count+stop_count
total_other_count=overall_other_count+other_count
```

```python
overall_count=total_positive_count+total_negative_count+total_stop_count+total_other_count
```

Considering the word trump to not be pos,neg or stop and then calculating the updated pos-neg ratio

```python
new_pos_count=0
new_neg_count=0
for word in words:
    if word == "trump":
        continue
    if word in positive_words:
        new_pos_count+=1
    if word in negative_words:
        new_neg_count+=1     
new_posnegratio=new_pos_count/new_neg_count # Updated pos to neg ratio
```

Calculating the positive and negative ratio

```python
positive_ratio=total_positive_count/overall_count
negative_ratio=total_negative_count/overall_count
pos_neg_ratio=total_positive_count/total_negative_count
stop_count_ratio=total_stop_count/overall_count
```

Using if else setting sentiment, based on the count of pos and neg words. So if pos words are more then the sentiment is positive and vice versa

```python
if total_positive_count > total_negative_count:
    overall_sentiment = "Positive"
elif total_negative_count > total_positive_count:
    overall_sentiment = "Negative"
else:
    overall_sentiment = "Neutral"
```

```python
print("Total Positive Count:", total_positive_count)
print("Total Negative Count:", total_negative_count)
print("Total Stop Count:", total_stop_count)
print("Total Other Count:", total_other_count)
print("Total Count:", overall_count)
print("Positive Ratio:", positive_ratio)
print("Negative Ratio:", negative_ratio)
print("Stop Count Ratio:", stop_count_ratio)
print("Other word ratio:", total_other_count/overall_count)
print("Positive/Negative Ratio:", pos_neg_ratio)
print("Overall Sentiment:", overall_sentiment)
print("updated pos/neg ratio:",new_posnegratio)
```
